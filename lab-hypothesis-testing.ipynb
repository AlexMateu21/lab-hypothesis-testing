In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
print("H0: AverageHP of Type Dragon <= Type Grass")
print("Ha: AverageHP of Type Dragon > Type Grass")
print("")
dragon_hp = df[( df['Type 1'] == 'Dragon') | (df['Type 2'] == 'Dragon')]['HP']  # In case Dragon appeared in type 1 or 2, same with grass
grass_hp  = df[(df['Type 1'] == 'Grass')  | (df['Type 2'] == 'Grass')]['HP']

#Two-sample t-test, one-tailed (alternative='greater' = Dragon > Grass)
t_stat, p_value_two = st.ttest_ind(dragon_hp, grass_hp)
# P_value / 2 since is two tailed
p_value_one = p_value_two / 2

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value (one-tailed): {p_value_one:.4f}")

if p_value_one < 0.05:
    print("We reject the null hypothesis, so Dragons have > HP than grass")
else:
    print("We accept the alternative hypothesis, so Dragons have <= than grass")

print("")
print(f"HP average of Dragon: {dragon_hp.mean():.2f}")
print(f"HP average of Grass: {grass_hp.mean():.2f}")
```

The P-value is almost 0, we can conclude that Dragon type have more HP than grass type

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
stats_pok = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

legendary = df[df["Legendary"] == True]
non_legendary = df[df["Legendary"] == False]

for stat in stats_pok:
    t_stat, p_value = st.ttest_ind(legendary[stat], non_legendary[stat])
    result = "Reject H0" if p_value < 0.05 else "Accept Ha"
    print(f"{stat}: t={t_stat:.4f}, p={p_value:.4f} → {result}")
    print(f"Legendary mean: {legendary[stat].mean():.2f} | Non-Legendary mean: {non_legendary[stat].mean():.2f}")
    print("")
```

The P_value is close to 0, obviously the Legenday Pokemons have different stats, much stronger in every aspect since they are unique and extremly powerful

```python
print("H0: The houses near a school or a hospital are not more expensive")
print("Ha: The houses near a school or a hospital are more expensive")
print("")
def euclidean_distance(long, lat, ref_long, ref_lat):
    return np.sqrt((long-ref_long)**2 + (lat-ref_lat)**2)

df["dist_school"]   = euclidean_distance(df["longitude"], df["latitude"], -118, 34)
df["dist_hospital"] = euclidean_distance(df["longitude"], df["latitude"], -122, 37)

close = df[(df['dist_school'] < 0.5) | (df['dist_hospital'] < 0.5)]['median_house_value']
far = df[(df['dist_school'] >= 0.5) & (df['dist_hospital'] >= 0.5)]['median_house_value']

print(f"Close houses: {len(close)}")
print(f"Far houses: {len(far)}")

t_stat, p_two = st.ttest_ind(close, far)
p_one = p_two / 2
print(f"\nT-statistic: {t_stat:.4f}")
print(f"P-value (one-tailed): {p_one:.4f}")
print("")
if p_one < 0.05 and t_stat > 0:
    print("We reject the H0, which means that The houses near a school or a hospital are more expensive.")
else:
    print("We accept the Ha, which means that The houses near a school or a hospital are not more expensive.")
```

The p-value is practically 0, we can say that houses near a school or a hospital are more expensive and this doesn't happen by chance
